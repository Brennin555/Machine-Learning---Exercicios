```python
#verifica se existe arquivo converted.wav e apaga caso tenha
if os.path.exists("/content/converted.wav"):
    os.remove("/content/converted.wav")
```

```python
# Caminho fixo do arquivo de áudio
filename = "/content/joke.mp3"

recognizer = sr.Recognizer()

# Converte para WAV se não estiver nesse formato
if not filename.endswith(".wav"):
    audio = AudioSegment.from_file(filename)
    filename = "converted.wav"
    audio.export(filename, format="wav")

with sr.AudioFile(filename) as source:
    audio_data = recognizer.record(source)
    try:
        text = recognizer.recognize_google(audio_data)
        print("Texto reconhecido:", text)
        textoReconhecido = text.lower()
    except sr.UnknownValueError:
        print("Não foi possível entender o áudio.")
        textoReconhecido = ""
    except sr.RequestError:
        print("Erro no serviço de reconhecimento de fala.")
        textoReconhecido = ""
```

```python
Audio("converted.wav", autoplay=True)
```

```python
# Função para falar
def speak(text):
    tts = gTTS(text=text, lang='en')
    filename = "/content/answer.mp3"
    tts.save(filename)
    return Audio(filename, autoplay=True)

# Função para processar comandos

if 'joke' in text:
    response = pyjokes.get_joke()

elif 'what time' in text:
    response = datetime.today().strftime("%H:%M %p")

elif 'exit' in text:
    response = "Goodbye!"

else:
    response = "Comando não reconhecido."

resposta = response  # Sempre gera áudio
```

```python
# ** Processa o áudio automaticamente **
recognized_text = textoReconhecido
if recognized_text:
    speak(resposta)

Audio("answer.mp3", autoplay=True)
```
